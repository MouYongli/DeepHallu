```python
from datasets import load_dataset

# Load LLaVA-Bench-in-the-Wild dataset
llava_bench = load_dataset("lmms-lab/llava-bench-in-the-wild")

# Print basic dataset info
print("Dataset structure:", llava_bench)
print("\nNumber of examples:", len(llava_bench['train']))

# Show a sample
sample = llava_bench['train'][0]
print("\nSample example:")
print("Image_id:", sample['image_id'])
print("Question_id:", sample['question_id'])
print("Question:", sample['question'])
print("Caption:", sample['caption']) 
print("GPT_answer:", sample['gpt_answer'])
print("Category:", sample['category'])


import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
plt.imshow(sample['image'])
plt.show()
```

```python
for i, sample in enumerate(llava_bench["train"]):
    print(f'Image ID: {sample["image_id"]}, Question ID: {sample["question_id"]}')
    print(f'\tQuestion: {sample["question"]}')
    print(f'\tCaption: {sample["caption"]}')
    print(f'\tGPT_answer: {sample["gpt_answer"]}')
    print(f'\tCategory: {sample["category"]}')
```
